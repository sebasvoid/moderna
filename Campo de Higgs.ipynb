```python
import matplotlib.pyplot as plt
import numpy as np
import platform
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

xvec = np.linspace( -5.75, 5.75, 50 )
yvec = np.linspace( -5.75, 5.75, 50 )
xmat, ymat = np.meshgrid ( xvec, yvec )
rmat = np.sqrt ( xmat**2 + ymat**2 )
zmat = np.sin(rmat) / rmat 

fig = plt.figure ( )

ax = fig.add_subplot ( 111, projection='3d' )
ax.plot_surface ( xmat, ymat, zmat, cmap = cm.coolwarm, edgecolor = 'none')
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
filename = 'Campo de Higgs.png'
plt.savefig ( filename )
plt.show ( )
```
